```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
%matplotlib inline

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.figure(figsize=(9, 10))
plt.plot(x, y)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

# using the fontdict property/attr. to set the fonts of any text eg title and x&y labels 
# the fontdict property collects a dictionary 
#  the loc attr. to set the positon of the title in the title( loc="positon") to align the title
plt.title("matplotlib practice", fontdict=font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 10))

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.subplot(1, 2, 1)
plt.plot(x, y)

x1 = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y1 = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.subplot(1, 2, 2)
plt.plot(x1, y1)

plt.title("matplotlib practice")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
# use the grid() to apply a grid layout on the matplotlib graph 
# axis attr./ args to display the layout of the lines "x", "y" or "BOTH"(default)
# apply all line decorations u learned 
plt.grid(axis="x", c='g', ls=":", lw=1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# use the subplot to display mutiple plots on the same figure
# The subplot() function takes three arguments that describes the layout of the figure.

# The layout is organized in rows and columns, which are represented by the first and second argument.

# The third argument represents the index of the current plot.
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 2, 2)
plt.plot(x,y)

# plot3
x = np.array([5, 1, 9, 3])
y = np.array([10, 40, 10, 70])

plt.subplot(3, 2, 3)
plt.plot(x,y)

#plot 4:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 2, 4)
plt.plot(x,y)

#plot 5:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 2, 5)
plt.plot(x,y)
plt.show()
```
