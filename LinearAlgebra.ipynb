```python
import numpy as np
import matplotlib.pyplot as plt

def picard_iteration(f, x0, y0, n_iter):
    # f: function defining the ODE dy/dx
    # x0: initial x value
    # y0: initial y value
    # n_iter: number of iterations
    
    x = np.linspace(x0, x0 + n_iter, n_iter + 1)
    y = np.zeros_like(x)
    y[0] = y0
    
    for i in range(n_iter):
        y[i + 1] = y0 + f(x[i], y[i])
    
    return x, y

def f(x, y):
    return x**2 + y

x0 = 0
y0 = 1
n_iter = 10

x, y = picard_iteration(f, x0, y0, n_iter)

plt.plot(x, y, 'b-o')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution of dy/dx = x^2 + y using Picard's method")
plt.grid(True)
plt.show()
```
