```python
def main():
    string = input("Enter a string: ").strip()
    result =""
    print(string)
    for ch in string:
        if ch.isalpha():
            result += str(getNumber(ch.upper()))
        else:
            result +=ch
    print(result)

def getNumber(uppercaseLetter):
    if uppercaseLetter.__ge__("W"):
        return 9
    elif uppercaseLetter.__ge__("T"):
        return 8
    elif uppercaseLetter.__ge__("P"):
        return 7
    elif uppercaseLetter.__ge__("M"):
        return 6
    elif uppercaseLetter.__ge__("J"):
        return 5
    elif uppercaseLetter.__ge__("G"):
        return 4
    elif uppercaseLetter.__ge__("D"):
        return 3
    else:
        return 2

main()
```
