# linear Algebra


```python
import numpy as np
from numpy.linalg import norm

v1 = np.array([[2, 3, 4]])
v2 = np.array([[2], [3], [4]])
```

```python
print(norm(v2, 2))
```

```python
print(np.cross(v1, v2.T))
```

```python
print(np.eye(2))
```

```python
from numpy.linalg import det

print(det(np.array([
    [1, 2, 4], 
    [2, 5, 3],
    [1, 7, 8]
])))
```

```python
from numpy.linalg import inv
z = np.array([
    [1, 2, 4], 
    [2, 5, 3],
    [1, 7, 8]
])
print(inv(z))
```
