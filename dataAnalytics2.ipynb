```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib notebook

φ = np.linspace(0, 2*np.pi, 1000)
r = 3

x = r*(φ - np.sin(φ))
y = r*(1 - np.cos(φ))

plt.figure(figsize=(8, 6))

plt.plot(x, y, label="cubic")
# plt.xlim(0, 20)
# plt.ylim(-6, 6)
plt.xlabel("X",family="courier", fontsize=30)
plt.ylabel("Y", fontfamily="san-serif")
plt.legend()
plt.show()
```
