```python
import numpy as np
import time

# Generate a simple signal of length 2048
N = 2048
n = np.arange(2048)
freq = 50
simple_signal = 2*np.sin(2*np.pi*50*n/N) + 5*np.sin(2*np.pi*100*n/N)

# Compute the FFT and record the time
start_time = time.time()
fft_result = np.fft.fft(simple_signal)
end_time = time.time()
fft_time = end_time - start_time

# Compute the DFT and record the time
start_time = time.time()
dft_result = np.zeros_like(simple_signal, dtype=np.complex)
for k in range(len(simple_signal)):
    for n in range(len(simple_signal)):
        dft_result[k] += simple_signal[n] * np.exp(-2j*np.pi*k*n/len(simple_signal))
end_time = time.time()
dft_time = end_time - start_time

print("FFT time:", fft_time)
print("DFT time:", dft_time)
#run on ur jupiter notebook.
```

```python
import numpy as np

def my_cubic_spline(x, y, X):
    n = len(x)
    A = np.zeros((n, n))
    b = np.zeros(n)
    c = np.zeros(n)
    d = np.zeros(n)

    # Compute the coefficients of the natural cubic spline
    h = np.diff(x)
    alpha = np.zeros(n-2)
    for i in range(1, n-1):
        alpha[i-1] = 3/h[i]*(y[i+1]-y[i]) - 3/h[i-1]*(y[i]-y[i-1])
    l = np.zeros(n)
    mu = np.zeros(n)
    z = np.zeros(n)
    l[0] = 1
    mu[0] = z[0] = 0
    for i in range(1, n-1):
        l[i] = 2*(x[i+1]-x[i-1])-h[i-1]*mu[i-1]
        mu[i] = h[i]/l[i]
        z[i] = (alpha[i-1]-h[i-1]*z[i-1])/l[i]
    l[n-1] = 1
    z[n-1] = c[n-1] = 0
    for j in range(n-2, -1, -1):
        c[j] = z[j] - mu[j]*c[j+1]
        b[j] = (y[j+1]-y[j])/h[j] - h[j]*(c[j+1]+2*c[j])/3
        d[j] = (c[j+1]-c[j])/(3*h[j])

    # Evaluate the cubic spline at the given points
    Y = np.zeros(len(X))
    for i in range(len(X)):
        j = np.searchsorted(x, X[i])
        if j == 0:
            j = 1
        elif j == n:
            j = n-1
        t = (X[i]-x[j-1])/h[j-1]
        Y[i] = y[j-1] + b[j-1]*t + c[j-1]*t**2 + d[j-1]*t**3

    return Y

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 1, 3, 2, 1])
X = np.linspace(1, 5, 50)
Y = my_cubic_spline(x, y, X)

import matplotlib.pyplot as plt
plt.plot(x, y, 'o', label='data')
plt.plot(X, Y, label='cubic spline')
plt.legend()
plt.savefig("assignment2outputresult.pdf", format="pdf")
plt.show()
```
